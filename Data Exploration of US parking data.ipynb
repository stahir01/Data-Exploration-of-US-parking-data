We will check how many Total Parking Spaces	 are owned/leased and what are those building status

```python
plt.bar(parking_data['Owned/Leased'], parking_data['Total Parking Spaces'])
plt.xlabel('Owned/Leased')
plt.ylabel('Total Parking Space')
```

## In which building was the parking space in excess?

```python
excess_parking = pd.pivot_table(parking_data, index= 'Bldg State', columns= 'Bldg Status', values = 'Total Parking Spaces')
excess_parking
```

```python
excess_parking = excess_parking[excess_parking['EXCESS'] > 0]
excess_parking
```

```python
excess_parking_2 = excess_parking.drop(columns=['ACTIVE', 'DECOMMISSIONED'])
excess_parking_2
```

```python
excess_parking_2.plot(kind='bar')
plt.ylabel('Amount in Excess')
```

```python
excess_parking_2 = excess_parking_2.sort_values('EXCESS', ascending=False)
excess_parking_2.plot(kind='bar')
plt.ylabel('Amount in Excess')
```

## In which type of property is parking space in excess

```python
prop_park_excess = parking_data.drop(columns = ['Bldg City', 'Bldg State', 'Owned/Leased'])
prop_park_excess =pd.pivot_table(prop_park_excess[prop_park_excess['Bldg Status'] == 'EXCESS'], index= 'Property Type',values = 'Total Parking Spaces')
prop_park_excess 
```

# States with more leased parking

```python
states_leased_parking = pd.pivot_table(parking_data[parking_data['Owned/Leased'] == 'LEASED'], index='Bldg State', values = 'Total Parking Spaces')
states_leased_parking = states_leased_parking.sort_values('Total Parking Spaces', ascending=False)
states_leased_parking.head()
```
